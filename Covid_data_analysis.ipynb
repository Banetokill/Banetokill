```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
infected = pd.read_csv("covid_19_india.csv")
infected.head()
```

```python
infected = infected.drop(['ConfirmedForeignNational','ConfirmedIndianNational'], axis =1 )
```

```python
infected.head()
```

```python
nan = infected.isnull().values.any()
print(nan)
```

```python
vaccine = pd.read_csv("covid_vaccine_statewise.csv")
vaccine.head()
```

```python
nan_1 = vaccine.isnull().values.any()
print(nan_1)
```

```python
vaccine = vaccine.fillna(0)
vaccine.head()
```

```python
df = pd.concat([infected,vaccine],axis = 0)
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df.to_csv('file.csv')
```

```python
df = df.drop(['Time'],axis =1)
df.head()
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df.head()
```

```python
df_s = df.sum(axis = 0, skipna = True)
df_s
```

## Total number of active cases :-

```python
df_active_cases = df_s['Confirmed']-df_s['Cured']-df_s['Deaths']
df_active_cases
```

```python
infected['active case'] = infected['Confirmed']-infected['Cured']-infected['Deaths']
infected.tail()
```

```python
statewise = pd.pivot_table(infected, values=['Confirmed','Cured','Deaths'],index ='State/UnionTerritory',aggfunc = max)
statewise
```

```python
statewise['recovery'] = statewise['Cured']*100/statewise['Confirmed']
statewise['Mortality'] = statewise['Deaths']*100/statewise['Confirmed']
statewise
```

```python
infected = infected.sort_values(by=['active case'], ascending=False)
infected.head()
```

```python
state_active = pd.pivot_table(infected,values = ['active case'],index = 'State/UnionTerritory',aggfunc = max)
state_active = state_active.sort_values(by=['active case'], ascending=False)
state_active.head(10)
```

```python
activegroup = infected.groupby(by = 'State/UnionTerritory').max()[['active case','Date']].sort_values(by =['active case'],ascending = False).reset_index()
activegroup
```

## Top 5 states with most number of active cases :- 

```python
x = activegroup['State/UnionTerritory'].head(10)
y = activegroup['active case'].head(10)
plt.bar(x, y)
plt.xticks(rotation = 90)
plt.figsize=(30,20)
plt.show()
```

```python
activegroup = infected.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by =['Deaths'],ascending = False).reset_index()
activegroup
```

## Top 5 states with highest deaths :-

```python
x = activegroup['State/UnionTerritory'].head(7)
y = activegroup['Deaths'].head(7)
plt.bar(x, y)
plt.xticks(rotation = 90)
plt.figsize=(30,20)
plt.xlabel('States/Union Territory')
plt.ylabel('Deaths')
plt.show()
```

```python
import seaborn as sns
# hue by states
# style to hue
fig = plt.figure(figsize = (100,50))
sns.lineplot(x= 'Date',data= infected[infected['State/UnionTerritory'].isin(['Maharashtra','Kerla','Karnataka','Tamil Nadu'])], y= 'active case',hue="State/UnionTerritory", style="State/UnionTerritory")
plt.xticks(rotation = 30)
plt.show()
```

```python
df = vaccine[vaccine["State"].str.contains("India")==False]
df
```

```python
pivot = df.pivot_table(index =['State'],
                       values =['Total Doses Administered'],
                       aggfunc ='sum')
max_sort = pivot.sort_values('Total Doses Administered',ascending = False)[:5]
max_sort.reset_index(inplace=True)
max_sort
```

## Top 5 states with highest number of doses administered :-

```python
v = max_sort['Total Doses Administered']
plt.bar(max_sort['State'],v)
plt.xlabel('State')
plt.ylabel('Doses administered')
plt.show()
```

```python
min_sort = pivot.sort_values('Total Doses Administered',ascending = True)[:5]
min_sort.reset_index(inplace=True)
min_sort
```

## Bottom 5 states with lowest number of doses administered :-

```python
v_m = min_sort['Total Doses Administered']
plt.bar(min_sort['State'],v_m)
plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('Doses administered')
plt.show()
```
