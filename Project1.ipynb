```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Downloads\HEA Phase DataSet v1d.csv", encoding="latin1")
df
```

```python
df.isna().sum()
```

```python
df_clean = df.dropna(subset=["Alloy ID"])
df_clean.isna().sum()
```

```python
df_clean
```

```python
DF1 = df_clean.drop(["References","Unnamed: 51","Unnamed: 52"],axis = 1)
DF1
```

```python
DF1 = DF1.dropna(subset = ["Density_calc"])
```

```python
DF1.isna().sum()
```

```python
X = DF1[["Density_calc","dHmix","dSmix","dGmix"]]

Y = DF1["Phases"]
```

```python
print("Before split:")
print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print("\nAfter split:")
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression(class_weight='balanced', max_iter=10000)
Lr.fit(X_train,Y_train)
```

```python
Pred = Lr.predict(X_test)
Pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(Pred,Y_test))
print(confusion_matrix(Pred,Y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train, Y_train)
```

```python
P1 = model.predict(X_test)
P1
```

```python
print(classification_report(P1,Y_test))
print(confusion_matrix(P1,Y_test))
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
cm = confusion_matrix(Y_test, Pred)
cm1 = confusion_matrix(Y_test, P1)
# Plotting confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
report = classification_report(Y_test, P1, output_dict=True)

# Convert to DataFrame
report_df = pd.DataFrame(report).transpose()

# Plot Precision, Recall, F1-Score for each class
report_df.drop('accuracy', axis=0, inplace=True)  # Remove accuracy row

fig, ax = plt.subplots(figsize=(10, 6))
report_df[['precision', 'recall', 'f1-score']].plot(kind='bar', ax=ax)

# Customize plot
plt.title('Classification Report')
plt.xlabel('Class')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(title='Metrics')

plt.tight_layout()
plt.show()
```
